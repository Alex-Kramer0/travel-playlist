```python
# imports
import pandas as pd
import os
```

Assumption: that all of the city data is located in the a folder named 'Extracts' within the same directory 


```python
extracts_path = os.path.join(os.getcwd(), "Extracts")
```

```python
columns_to_keep = [
    "id",
    "listing_url",
    "name",
    "description",
    "neighborhood_overview",
    "neighbourhood_cleansed",
    "property_type",
    "room_type",
    "accommodates",
    "amenities" 
]
```

```python
# for each file in the 'Extracts' folder 

dfs = []


for file in os.listdir(extracts_path):
    if file.lower().endswith(".csv"):
        file_path = os.path.join(extracts_path, file)
        df = pd.read_csv(file_path)

        # Keep only columns that exist in the current file
        df = df[[col for col in columns_to_keep if col in df.columns]]

        dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True, sort=False)

output_path = os.path.join(os.getcwd(), "Output.csv")
combined_df.to_csv(output_path, index=False)

print(f"Combined {len(dfs)} files into {output_path}")
```
